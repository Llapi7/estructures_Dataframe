```python

#Resumeix estadísticament el dataset i les columnes d’interès. Fes una anàlisi estadístic del que consideris rellevant.
#Troba quantes dades faltants hi ha per columna.
#Crea columnes noves (velocitat mitjana del vol, si ha arribat tard o no...).
#Fes una taula de les aerolínies amb més endarreriments acumulats.
#Quins són els vols més llargs? I els més endarrerits? Busca les rutes més llargues i les que acumulen més retards.
#Aporta allò que consideris rellevant.

import numpy as np
import pandas as pd

na_values = ['NaN','0.0']
flights= pd.read_csv("DelayedFlights.csv")

df2 = pd.DataFrame(flights)
df1 = df2.sample(n = 200000, random_state =1)
df1.head()
```

```python
np.shape(df1)
```

```python
#Troba quantes dades faltants hi ha per columna.

df1.isnull().sum(axis=0)
```

```python
# Muchos Arrtime nulos...?


```

```python
x = df1["ArrTime"].median()
y = df1["ActualElapsedTime"].median()
z = df1["CRSElapsedTime"].median()
xx= df1["AirTime"].median()
yy= df1["ArrDelay"].median()



df1["ArrTime"].fillna(x, inplace = True)
df1["ActualElapsedTime"].fillna(y, inplace = True)
df1["CRSElapsedTime"].fillna(z, inplace = True)
df1["AirTime"].fillna(xx, inplace = True)
df1["ArrDelay"].fillna(yy, inplace = True)
df1["CarrierDelay"].fillna(0, inplace = True)


df1.isnull().sum(axis=0)
```

```python
df1["NASDelay"].fillna(0, inplace = True)
df1["SecurityDelay"].fillna(0, inplace = True)
df1["LateAircraftDelay"].fillna(0, inplace = True)
#df1["WeatherDelay"].fillna(0, inplace = True)



df1.isnull().sum(axis=0)
```

```python
df1.dropna(subset=['TailNum','TaxiOut','TaxiIn'], inplace = True)
df1.isnull().sum(axis=0)
```

```python
np.shape(df1)
```

```python
df3 = df1.drop(["DepTime","CRSDepTime", "Unnamed: 0","FlightNum","Year","Month","DayofMonth","DayOfWeek", "TailNum", "CancellationCode", "Cancelled", "Diverted"],axis = 'columns')


df3.drop_duplicates(inplace = True)
```

```python
df3.isnull().sum(axis=0)
#df3.loc["AirTime"=='Nan']
```

```python
np.shape(df3)
df3.head()
```

```python
#  new DF with less columns
df3.describe()
```

```python
Suma_Valors = df3.sum( axis = 0 , skipna = True)
Suma_Valors
```

```python
Total_time  = df3['AirTime'].sum()


print("El temps total de vol es de " + str(Total_time)+ " minuts, és a dir "+(str(Total_time/60))+"hores")
      


    
Total_distance = df3['Distance'].sum()



Velocitatmitjana =  round(Total_distance/ (Total_time)*60, 2)
print("La Velocitat mitjana és de " + (str((Velocitatmitjana)))+ " M/Hora")
```

```python
# Noves Columnnes:

                  # velocitat-mitjana

Total_distance = round(df3["Distance"], 2)
Total_distance

Total_time = round(df3['AirTime'],2)
Total_time

df3["Velocitat_mitjana"] =round(Total_distance/Total_time, 2)



```

```python
                # Retard ?
    
Total_delay = round(df3["ArrTime"] - df3["CRSArrTime"])
df3["Retard"] =round( Total_delay, 2)

```

Fes una taula de les aerolínies amb més endarreriments acumulats.



```python
#Idem pero con i.loc

new_df1 = df3.loc[:,['UniqueCarrier','Retard','Velocitat_mitjana']]

new_df1.head()

```

```python
#UniqueCarrier_delays = new_df1.sort_values(by=['Retard'], ascending = False)
UniqueCarrier_delays = new_df1.groupby('UniqueCarrier')['Retard'].sum().sort_values(ascending = True)
UniqueCarrier_delays
#UniqueCarrier_delays_sorted = UniqueCarrier_delays.sort_values(by ='Retard', ascending = False)
#UniqueCarrier_delays_sorted
```

```python
#Quins són els vols més llargs?  Busca les rutes més llargues i les que acumulen més retards.

Milles_Vol_més_llarg = df3['Distance'].max()


nº_Vol_més_llarg2  = df3['Distance'].idxmax()


print(str(Milles_Vol_més_llarg) + " Milles")
print("El vol més llarg és el nº : "+ str(nº_Vol_més_llarg2))



#Vol_més_llarg3 = df3.where[(df3.id) == 1068823]



#Vol_més_llarg = df3.iloc[['Distance'].idxmax(),('Origin')]


```

```python
# Rutes més llargues 

Origin = df3.groupby('Origin') # Agrupo df por Origen


Rutes_més_llargues = Origin[('Distance')]
Rutes_més_llargues.max().sort_values(ascending = False)



```

```python
# Rutes amb més retard

Rutes_més_retard = Origin[('CRSElapsedTime')]
Rutes_més_retard.max().sort_values(ascending = False)
```

```python
#I els més endarrerits?

Origin = df3.groupby('Origin') 
Vol_més_endarrerit = Origin[('CRSElapsedTime')]
Vol_més_endarrerit.max().sort_values(ascending = False)


```
